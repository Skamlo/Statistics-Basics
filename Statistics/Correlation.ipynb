# Covariance

Sources:
- <a href="https://www.youtube.com/watch?v=xZ_z8KWkhXE">StatQuest - Covariance</a>

$$
Cov(X, y) = \frac{1}{m} \sum_{i=1}^{m} \left( x_i - \bar{x} \right)\left( y_i - \bar{y} \right)
$$

```python
np.cov(
    data[["horsepower", "weight"]].to_numpy().T
)
```

# Pearson's Correlation

Sources:
- <a href="https://www.youtube.com/watch?v=xZ_z8KWkhXE">StatQuest - Pearson's Correlation</a>

$$
Corr(X, Y) = \frac{ Cov(X, Y) }{ \sigma_X \cdot \sigma_Y }
= \frac{ \sum \left( x_i - \bar{x} \right)\left( y_i - \bar{y} \right) }
{\sqrt{ \sum \left( x_i - \bar{x} \right)^2 \sum \left( y_i - \bar{y} \right)^2 }}
$$

```python
data[["horsepower", "weight"]].corr()
```

# Spearman's Correlation

Spearman's correlation is used when the relationship between the two variables are non-linear.

Soruces:
- <a href="https://www.youtube.com/watch?v=XV_W1w4Nwoc">DATAtab - Spearman Rank Correlation</a>
- <a href="https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient">Wikipedia</a>

For a sample of size m, the n raw scores $X_i$, $Y_i$ are converted to ranks $R(X_i)$, $R(Y_i)$

<br>

$$
Corr(X, y) = \frac{ Cov(R(X), R(Y)) }{ \sigma_{R(X)} \cdot \sigma_{R(Y)} }
= 1 - \frac{6 \sum \left( R(X_i) - R(Y_i) \right)^2 }{ n(n^2 -1) }
$$

```python
data[["horsepower", "weight"]].corr(method="spearman")
```
