# Data

```python
def rotate(point, angle):
    qx = np.cos(angle) * point[0] - np.sin(angle) * point[1]
    qy = np.sin(angle) * point[0] + np.cos(angle) * point[1]
    return qx, qy

X = np.array([
    np.random.normal(size=1000, scale=2),
    np.random.normal(size=1000, scale=0.5)
]).T

for i in range(X.shape[0]):
    qx, qy = rotate(X[i], -np.pi/4)
    X[i][0], X[i][1] = qx, qy

plt.figure(figsize=(8, 6))
plt.scatter(X.T[0], X.T[1], color="darkblue")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

# Multivariate Normal Distribution

$$
\frac{1}{\left(2 \pi\right)^\frac{d}{2} |\Sigma|^\frac{1}{2}}
exp\left( -\frac{1}{2}X^TX \right)
$$

$$
\frac{1}{\left(2 \pi\right)^\frac{d}{2} |\Sigma|^\frac{1}{2}}
exp\left( -\frac{1}{2}\left(X-\mu\right)^T \Sigma^{-1} \left(X-\mu\right) \right)
$$

$$
\Sigma = \left[ \begin{array}{cc}
    D^2(X) & Cov(X, Y) \\
    Cov(Y, X) & D^2{Y}
\end{array} \right]

\hspace{2em}

X = \left[ \begin{array}{c}
    x_1 \\
    x_2
\end{array} \right]

\hspace{2em}

\mu = \left[ \begin{array}{c}
    \mu_1 \\
    \mu_2
\end{array} \right]
$$
