# Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Binomial distribution

Describe a number of $k$ successes in sequence of $n$ independent experiments.

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Binomial_distribution">Wikipedia</a>
- <a href="https://www.youtube.com/watch?v=8idr1WZ1A7Q">3Blue1Brown</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html">Scipy DOCS</a>

$$
X \sim B(n, p)
$$

$$
P(A) = \left( \begin{array}{c} n \\ k \end{array} \right) \cdot p^k \cdot (1-p)^{n-k}
$$

$$
E(X)=np \hspace{3em} Var(X)=np(1-p)
$$

```python
n = 50
p = 0.3
distribution = [stats.binom.pmf(k, n, p) for k in range(n + 1)]

plt.figure(figsize=(12, 3))
for i, prob in enumerate(distribution):
    plt.bar(i, prob, width=1, color="#ff8819")
plt.title(f"Binomial distribution   p={p}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Negative binomial distribution

Negative binomial distribution talks about the final success which can be obtained, after a sequence of successes in the preceding trials.

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Negative_binomial_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html">Scipy DOCS</a>

$$
X \sim NB(r, p)
$$

$$
P(A) = \left( \begin{array}{c} r+k-1 \\ r \end{array} \right) \cdot p^r \cdot (1-p)^k
$$

$$
E(X)=\frac{rp}{1-p} \hspace{3em} Var(X)=\frac{rp}{(1-p)^2}
$$

```python
n = 50
p = 0.6
distribution = [stats.nbinom.pmf(k, n, p) for k in range(n + 1)]

plt.figure(figsize=(12, 3))
for i, prob in enumerate(distribution):
    plt.bar(i, prob, width=1, color="#ff8819")
plt.title(f"Negative binomial distribution   p={p}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Geometric distribution

Describes the number of trials needed to achieve first success in a series of independent Bernoulli trials.

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Geometric_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.geom.html">Scipy DOCS</a>

$$
X \sim G(p)
$$

$$
P(A) = (1-p)^{k-1} \cdot p
$$

$$
E(X)=\frac{1}{p} \hspace{3em} Var(X)=\frac{1-p}{p^2}
$$

```python
n=50
p = 0.1
distribution = [stats.geom.pmf(k, p) for k in range(n + 1)]

plt.figure(figsize=(12, 3))
for i, prob in enumerate(distribution):
    plt.bar(i, prob, width=1, color="#ff8819")
plt.title(f"Geometric distribution   p={p}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Poisson distribution

Describes the number of events per unit of time, space, or other unit of measurement when these events occur independently of each other and with the same average frequency.

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Poisson_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html">Scipy DOCS</a>

$$
X \sim Pois(\lambda)
$$

$$
P(A) = \frac{\lambda^k}{k!}e^{-\lambda}
$$

$$
E(X)=\lambda \hspace{3em} Var(X)=\lambda
$$

```python
n = 50
lambda_param = 10
distribution = [stats.poisson.pmf(k, lambda_param) for k in range(n + 1)]

plt.figure(figsize=(12, 3))
for i, prob in enumerate(distribution):
    plt.bar(i, prob, width=1, color="#ff8819")
plt.title(r"Poisson distribution  $\lambda$=" + str(lambda_param))
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Continuous uniform distribution

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Continuous_uniform_distribution">Wikipedia</a>
- <a href="">Scipy DOCS</a>

$$
X \sim U\left[a, b\right]
$$

$$
f(x) = \Biggl\{
\begin{array}{cll}
    \frac{1}{b-a} & \text{for} & a \leq x \leq b, \\
    0 & \text{for} & x < a \hspace{0.5em} \text{or} \hspace{0.5em} x > b
\end{array}
$$

$$
E(X)=\frac{a+b}{2} \hspace{3em} Var(X)=\frac{(b-a)^2}{12}
$$

```python
n = 50
a, b = 20, 40
x = np.linspace(0, n, 1000)
distribution = stats.uniform.pdf(x, loc=a, scale=b-a)

plt.figure(figsize=(12, 3))
plt.plot(x, distribution, color="#ff8819")
plt.fill_between(x, [0]*(1000), distribution, color="#ffc187")
plt.title(f"Continuous uniform distribution  [a, b] = [{a}, {b}]")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Exponential distribution

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Exponential_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html">Scipy DOCS</a>

$$
X \sim Exp(\lambda)
$$

$$
f(x) = \Biggl\{
\begin{array}{cll}
    \lambda e^{-\lambda x} & \text{for} & x \geq 0, \\
    0 & \text{for} & x < 0
\end{array}
$$

$$
E(X)=\frac{1}{\lambda} \hspace{3em} Var(X)=\frac{1}{\lambda^2}
$$

```python
n = 50
lambda_param = 0.1
x = np.linspace(0, n, 1000)
distribution = stats.expon.pdf(x, scale=1/lambda_param)

plt.figure(figsize=(12, 3))
plt.plot(x, distribution, color="#ff8819")
plt.fill_between(x, [0]*(1000), distribution, color="#ffc187")
plt.title(r"Exponential distribution  $\lambda$=" + str(lambda_param))
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Normal distribution

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Normal_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html">Scipy DOCS</a>

$$
X \sim N(\mu, \sigma)
$$

$$
f(x) = \frac{1}{\sqrt{2 \pi} \sigma} e^{ - \large{ \frac{ \left( x - \mu \right)^2 }{ 2\sigma^2 } } }
$$

$$
E(X)=\mu \hspace{3em} Var(X)=\sigma^2
$$

```python
n = 50
mean = 25
std = 5
x = np.linspace(0, n, 1000)
distribution = stats.norm.pdf(x, loc=mean, scale=std)

plt.figure(figsize=(12, 3))
plt.plot(x, distribution, color="#ff8819")
plt.fill_between(x, [0]*(1000), distribution, color="#ffc187")
plt.title(rf"Normal distribution  $\mu$={mean}   $\sigma$={std}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Student's t-distribution

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Student%27s_t-distribution">Wikipedia</a>
- <a href="https://www.jarad.me/courses/stat587Eng/slides/Probability/Distributions/Students_t/Students_t.pdf">Iowa State University</a>
- <a href="https://statproofbook.github.io/P/t-pdf.html">The Book of Statistical Proofs</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html">Scipy DOCS</a>

$$
X \sim t_{\nu}
$$

$$
f(t) = \frac{ \Gamma \left( \frac{\nu + 1}{2} \right) }{ \sqrt{\pi \nu} \hspace{0.4em} \Gamma \left( \frac{\nu}{2} \right) }
\left( 1 + \frac{t^2}{\nu} \right)^{ -\large{ \frac{\nu + 1}{2} } }
\hspace{3em}
\Gamma(z) = \int_{0}^{\infty} t^{z-1} e^{-1} \hspace{0.25em} dt
$$

$$
E(X)=0 \hspace{3em} Var(X)=\frac{\nu}{\nu - 2}
$$

```python
n = 50
mean = 25
std = 5
df = 2
x = np.linspace(0, n, 1000)
distribution = stats.t.pdf(x, df=df, loc=mean, scale=std)

plt.figure(figsize=(12, 3))
plt.plot(x, distribution, color="#ff8819")
plt.fill_between(x, [0]*(1000), distribution, color="#ffc187")
plt.title(rf"Normal distribution  $\mu$={mean}   $\sigma$={std}   $df$={df}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```

# Chi-squared distribution

Sources:<br>
- <a href="https://en.wikipedia.org/wiki/Chi-squared_distribution">Wikipedia</a>
- <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html">Scipy DOCS</a>

$$
X \sim \chi_{k}^{2}
$$

$$
f(x) = \frac{1}{ \large{ 2^{\frac{k}{2}} } \hspace{0.25em} \Gamma \left( \frac{k}{2} \right) } \hspace{0.5em} \large{ x^{\frac{k}{2}-1} e^{-\frac{x}{2}} }
$$

$$
E(X)=k \hspace{3em} Var(X)=2k
$$

```python
n = 50
mean = 0
std = 5
df = 4
x = np.linspace(0, n, 1000)
distribution = stats.chi2.pdf(x, df=df, loc=mean, scale=std)

plt.figure(figsize=(12, 3))
plt.plot(x, distribution, color="#ff8819")
plt.fill_between(x, [0]*(1000), distribution, color="#ffc187")
plt.title(rf"Chi-squared distribution  $\mu$={mean}   $\sigma$={std}   $df$={df}")
plt.xticks(np.arange(0, 51, 5))
plt.show()
```
