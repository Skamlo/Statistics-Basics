# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def drawGraph(fn, name:str=None, color:str="orange"):
    x = np.linspace(-5, 5, 1000)
    y = fn(x)
    plt.figure(figsize=(10, 3))
    plt.plot(x, y, color=color)
    if name is not None:
        plt.title(name)
    plt.xlabel("x")
    plt.xticks(np.arange(-5, 6))
    plt.show()
```

# Step function

Sources:
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
f(x) = \Biggl\{
\begin{array}{cll}
    1 & \text{for} & x \geq 0, \\
    0 & \text{for} & x < 0
\end{array}
$$

```python
def step_function(x):
    return (x >= 0).astype("int8")

drawGraph(step_function, name="Step Function")
```

# Sigmoid

Sources:
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

drawGraph(sigmoid, name="Sigmoid")
```

# Softmax

Source:
- <a href="https://www.youtube.com/watch?v=KpKog-L9veg">StatQuest</a>
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
\sigma(\vec{z})_i = \large{ \frac{ e^{z_i} }{ \sum_{j=0}^{K} e^{z_j} } }
$$
<br>

$$ \text{Example:} $$
$$
\vec{v} = \left[\begin{array}{c} 1 \\ 2 \\ 3 \\ 4 \\ 5 \end{array}\right]
\hspace{1em} \longrightarrow \hspace{1em}
\vec{u} = \sigma(\vec{v})
\hspace{1em} \longrightarrow \hspace{1em}
\vec{u} = \left[\begin{array}{c} 0.01165623 \\ 0.03168492 \\ 0.08612854 \\ 0.23412166 \\ 0.63640865 \end{array}\right]
$$

```python
def softmax(x):
    denominator = np.sum(np.exp(x))
    new_x = np.exp(x) / denominator
    return new_x

x = np.array([1, 2, 3, 4, 5])
softmax(x)
```

# ReLU

Source:
- <a href="https://www.youtube.com/watch?v=68BZ5f7P94E">StatQuest</a>
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
\text{ReLU}(x) = max(0, x)
$$

```python
def ReLU(x):
    return np.maximum(0, x)

drawGraph(ReLU, name="ReLU")
```

# Leaky ReLU

Source:
- <a href="https://www.youtube.com/watch?v=cQ56gsYKIuc">StatQuest</a>
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
\text{LReLU}(x) = max(0.1x, x)
$$

```python
def LeakyReLU(x):
    return np.maximum(0.1*x, x)

drawGraph(LeakyReLU, name="Leaky ReLU")
```

# TanH

Sources:
- <a href="https://www.youtube.com/watch?v=Fu273ovPBmQ">AssemblyAI</a>

$$
\text{TanH}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} = \frac{2}{1 + e^{-2x}} - 1
$$

```python
def tanh(x):
    return 2 / (1 + np.exp(-2*x)) - 1

drawGraph(tanh, name="tanh")
```
